```python
## import standard libraries 

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
##read the csv files using pandas librabry with function name read_csv
dataset = pd.read_csv('CKD.csv')
```

```python
##read the stored value on the dataset
dataset
```

```python
##text to numerical values (0 & 1)--one hot encoding and avoid duplicates using drop_first
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
##read the stored value on the dataset
dataset
```

```python
##check wheather data is balance or imbalance
dataset["classification_yes"].value_counts()
```

```python
## input data variables
indep=dataset[["age", "bp", "al", "su", "bgr", "bu", "sc", "sod", "pot", "hrmo", "pcv",
       "wc", "rc", "sg_b", "sg_c", "sg_d", "sg_e", "rbc_normal", "pc_normal",
       "pcc_present", "ba_present", "htn_yes", "dm_yes", "cad_yes",
       "appet_yes", "pe_yes", "ane_yes"]]
```

```python
## output data variables
dep=dataset["classification_yes"]
```

```python
##split the training and test dataset -- to better understanding, we consider as X_train as model question paper/X_test as model paper answer
##y_train as final exam / y_test as final answer. 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
## standarisation method: mean = 0 and standard deviation = 1/ fit: learn the mean and standard only from training data/ scales the training data /
## use the same mean and standard learned from training data. Scales test data consistenly.
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
##using hypertuning parameter of the logisticRegression with the help of Grid search method to create the model.
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion':['gini', 'entropy', 'log_loss'], 'splitter':['best', 'random'] }
##refit=true -- finding the best parameter/n_jobs=-1 -- uses all cpu cores/f1_weighted -- good for imbalance dataset. it combines both recall and precision
##verbose=3 -- very detailed analysis 
grid = GridSearchCV(DecisionTreeClassifier(), param_grid, refit=True, n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,y_train)
```

```python
## cv_results_ detailed results of cross validation/stored as dictionary 
## use the best model found by Gridserachcv to verify the unseen test data
re=grid.cv_results_
grid_pred= grid.predict(X_test)
```

```python
##evaluation metics for classification
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, grid_pred)
```

```python
##evaluation metics for classification
from sklearn.metrics import classification_report
clf_report=classification_report(y_test, grid_pred)
```

```python
##calculates F1 score. compare actual vs predicted data. weighted: computes each class and good for imbalance dataset 
from sklearn.metrics import f1_score
f1_macro = f1_score(y_test, grid_pred, average='weighted')
print("the best f1_score parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print("The confusion Matrix :\n",cm)
```

```python
print("The classification report:\n",clf_report)
```

```python
##another evaluation metrics ROC_AUC : Returns probabilities, not class labels. [:,1 ]probability of class 1
##ROC–AUC must use probabilities or scores, not labels.
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, grid.predict_proba(X_test)[:,1])
```
